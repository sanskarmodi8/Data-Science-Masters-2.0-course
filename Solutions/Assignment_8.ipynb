```python
# Answer 3

#The ArithmeticError class in Python handles exceptions related to arithmetic operations. 
#Two examples of errors defined in the ArithmeticError class are --

#ZeroDivisionError --

try:
    a=24/0
except ZeroDivisionError as e:
    print("ERROR: {}".format(e))

#FloatingPointError --

import sys

def calculate_factorial(n):
    try:
        result = 1
        for i in range(1, n+1):
            result *= i
        print("Factorial of", n, ":", result)
    except OverflowError as e:
        print("Error:", e)

calculate_factorial(10**6)
```

```python
# Answer 4

#The LookupError class in Python is the base class for exceptions that occur when an index or key is not found in a sequence or mapping. 
#It serves as a superclass for specific lookup-related errors, such as KeyError and IndexError.

#example of keyerror --

my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']  # Accessing a non-existent key
except KeyError as e:
    print("KeyError:", e)

    
#example of indexerror --

my_list = [1, 2, 3]

try:
    value = my_list[3]  # Accessing an index beyond the list's length
except IndexError as e:
    print("Error:", e)
```
