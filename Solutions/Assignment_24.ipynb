```python
# Answer 1

# Five common plots that can be created using Seaborn and their typical uses:

# Scatter Plot (scatterplot): A scatter plot is used to visualize the relationship between two continuous variables. 
# It can help identify patterns, trends, or clusters in the data.

# Line Plot (lineplot): A line plot is used to display the relationship between two continuous variables over a continuous interval.
# It is often used to show trends or changes in data over time.

# Bar Plot (barplot): A bar plot is used to compare categorical data by showing the distribution or aggregate values for different categories.
# It is useful for visualizing and comparing values across categories.

# Histogram (histplot): A histogram is used to display the distribution of a single continuous variable. 
# It shows the frequency or count of data points falling into different bins or intervals.

# Heatmap (heatmap): A heatmap is used to visualize the relationship between two categorical variables by displaying a matrix of colors. 
# It is often used to show patterns or correlations in large datasets.
```

```python
# Answer 2

import seaborn as sns

data = sns.load_dataset('fmri')
sns.lineplot(x='timepoint', y='signal', data = data, hue='event', style='region')
```

```python
# Answer 3

data = sns.load_dataset('titanic')
sns.boxplot(x='pclass', y='age', data=data)
```

```python
sns.boxplot(x='pclass', y='fare', data=data)
```

```python
# Answer 4

data = sns.load_dataset('diamonds')
sns.histplot(x='price', hue='cut', data=data)
```

```python
# Answer 5

data = sns.load_dataset('iris')
sns.pairplot(data, hue='species')
```

```python
# Answer 6

data = sns.load_dataset('flights')
sns.heatmap(data=data.pivot("month", "year", "passengers"))
```
