```python
import plotly.express as px
import seaborn as sns
```

```python
# Answer 1

data = sns.load_dataset('titanic')
fig = px.scatter(data, x='age', y='fare')
fig.show()
```

```python
# Answer 2

data = sns.load_dataset('tips')
fig = px.box(data, x='day', y='total_bill', color='smoker' , notched=True)
fig.show()
```

```python
# Answer 3

fig = px.histogram(x=data['sex'], y=data['total_bill'], pattern_shape=data['smoker'], color=data['day'])
fig.show()
```

```python
# Answer 4

data = sns.load_dataset('iris')
fig = px.scatter_matrix(data, color='species', dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"])
fig.show()
```

```python
# Answer 5

# The distplot function is a visualization technique commonly used to display the distribution of a single numerical variable.
#  It combines a histogram with a kernel density estimate (KDE) plot, providing a comprehensive view of the data's distribution.

tips = sns.load_dataset('tips')
fig = px.histogram(tips, x="total_bill", y="tip", color="sex", marginal="rug",
                   hover_data=tips.columns)
fig.show()


```
