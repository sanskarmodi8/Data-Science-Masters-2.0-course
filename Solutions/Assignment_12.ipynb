```python
# Answer 1

#MongoDB is a popular document-oriented, NoSQL database

#Non-relational databases, or NoSQL databases, are designed to store and retrieve data without requiring a fixed schema
#NoSQL databases can be categorized into various types, including document databases (e.g., MongoDB), key-value stores, 
#columnar databases, and graph databases.

#MongoDB is preferred over SQL databases when handling unstructured or semi-structured data and when scalability, flexibility,
#and rapid development is required.
```

```python
# Answer 2

# There are many features of MongoDb from which, some of the most prominent of them are --

# Document-Oriented Storage: MongoDB stores data in a way that's similar to how you would organize information in a document or file. 
# It allows you to work with unstructured or changing data easily.

# Scalability and High Performance: MongoDB is designed to handle large amounts of data and high traffic. It can distribute the data 
# across multiple servers to handle a lot of users and ensure fast response times.

# Flexible Schema: In MongoDB, you don't have to define a fixed structure for your data upfront. You can easily change the data structure 
# as your needs evolve, which gives you more flexibility and makes development faster.
```
