````python
# Answer 2

# In Bokeh, glyphs are visual markers used to represent data points on a plot. They can be various shapes, such as circles, squares, triangles, lines, and more. Glyphs are added to a Bokeh plot to visually depict the data.

# To add glyphs to a Bokeh plot, you need to use the appropriate glyph function provided by Bokeh, such as circle(), square(), triangle(), line(), etc. These functions take the data coordinates and additional parameters to specify the appearance of the glyphs.

# example -- 

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_notebook()

p = figure(title="Bokeh Scatter Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

p.circle(x, y, size=10, color='navy', alpha=0.5)
p.square(x, y, size=8, color='red', alpha=0.8)

show(p)


````
