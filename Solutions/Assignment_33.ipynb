```python
# Answer 2

import numpy as np
import scipy.stats as stats

# Observed frequencies of colors in the bag
observed_freq = np.array([45, 55, 60, 30, 50, 60])  # Replace with actual observed frequencies

# Expected frequencies based on the given distribution
expected_freq = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed_freq)

# Perform the chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

# Compare p-value with significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The distribution of colors in the bag is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution of colors in the bag matches the expected distribution.")
```

```python
# Answer 4

import scipy.stats as stats

# Given data
sample_proportion = 60 / 500  # Proportion of smokers in the sample
sample_size = 500

# Calculate the standard error of the proportion
standard_error = (sample_proportion * (1 - sample_proportion) / sample_size)**0.5

# Calculate the 95% confidence interval
confidence_interval = stats.norm.interval(0.95, loc=sample_proportion, scale=standard_error)

print("95% Confidence Interval for the proportion of smokers:", confidence_interval)


# Interpretation:
# The 95% confidence interval indicates that we are 95% confident that the true proportion of smokers in the population lies within this range. In this case, the true proportion of smokers is likely to be between the lower bound and the upper bound of the confidence interval.

```

```python
# Answer 6

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Degrees of freedom
dof = 10

# Range of chi-square values for plotting
x = np.linspace(0, 30, 500)

# Calculate the chi-square distribution
chi2_distribution = stats.chi2.pdf(x, df=dof)

# Plot the chi-square distribution
plt.plot(x, chi2_distribution, label=f"Chi-square (df={dof})")
plt.axvline(x=15, color='red', linestyle='--', label="Chi-square statistic = 15")
plt.fill_between(x, chi2_distribution, 0, where=(x >= 15), color='red', alpha=0.3)

# Label axes and add legend
plt.xlabel("Chi-square Value")
plt.ylabel("Probability Density")
plt.legend()

# Show the plot
plt.show()
```

```python
# Answer 8

import scipy.stats as stats

# Observed frequencies
observed_tails = 45
observed_heads = 100 - observed_tails

# Expected frequencies for a fair coin
expected_tails = 50
expected_heads = 50

# Combine the observed and expected frequencies
observed_freq = [observed_tails, observed_heads]
expected_freq = [expected_tails, expected_heads]

# Perform the chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

# Compare p-value with significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The coin is biased towards tails.")
else:
    print("Fail to reject the null hypothesis. The coin appears to be fair.")
```

```python
# Answer 11

import scipy.stats as stats

# Given data
sample_mean = 72
sample_std = 10
sample_size = 30
population_mean = 70

# Calculate the standard error of the mean
standard_error = sample_std / (sample_size**0.5)

# Calculate the test statistic (t-score)
t_score = (sample_mean - population_mean) / standard_error

# Degrees of freedom for a one-sample t-test
dof = sample_size - 1

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(abs(t_score), df=dof))

# Compare p-value with significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the population mean and 70.")
```
