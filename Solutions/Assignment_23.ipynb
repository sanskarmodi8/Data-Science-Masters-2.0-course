```python
# Answer 1

# Matplotlib is a popular Python library used for creating visualizations, such as graphs and charts. It provides a comprehensive set of tools
# for creating high-quality plots in a variety of formats.
# The Pyplot module of Matplotlib allows for easy creation of different types of plots. 

# Five common plots that can be created using Pyplot are line plots, scatter plots, bar plots, histogram plots, and pie charts.
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
# Answer 2

# A scatter plot is a type of visualization that represents data points as individual markers on a two-dimensional coordinate system. 
# It is used to show the relationship or correlation between two variables

plt.scatter(x, y, c='r', alpha=.7)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Answer 2")
```

```python
# Answer 3

# The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple
# plots in separate axes, arranged in rows and columns, within a single figure window

x = np.array([0, 1, 2, 3, 4, 5]) 

plt.subplot(2,2,1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y1)
plt.title("Line 1")

plt.subplot(2,2,2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y2)
plt.title("Line 2")

plt.subplot(2,2,3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y3)
plt.title("Line 3")

plt.subplot(2,2,4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y4)
plt.title("Line 4")
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Answer 4

# A bar plot is a type of visualization that represents categorical data with rectangular bars. The length of each bar corresponds to the
# value of a specific category, allowing for easy comparison between categories.

plt.bar(company, profit)
```

```python

plt.barh(company, profit)
```

```python
# Answer 5

# A box plot, also known as a box-and-whisker plot, is a graphical representation of statistical data that displays the distribution of
# a dataset along with its summary statistics. It provides a visual summary of the minimum, first quartile, median, third quartile, and
# maximum values of a dataset, as well as any potential outliers.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
```
